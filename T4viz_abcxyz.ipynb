```python
import pandas as pd
```

```python
df4eda = pd.read_csv('df4eda.csv', index_col=0)
df4eda = df4eda.drop(['RunningMonths', 'Running Year', 'Running Months',
                     'Rolling Period', 'Rolling Period Sort'], axis = 1)
df4eda
```

```python
list_csv = ['df_abc_xyz_out__2014.csv', 'df_abc_xyz_out__2013.csv','df_abc_xyz_out__2012.csv',
           'df_abc_xyz_out__2011.csv', 'df_abc_xyz_out__2010.csv','df_abc_xyz_out__2009.csv',
           'df_abc_xyz_out__2008.csv', 'df_abc_xyz_out__2007.csv','df_abc_xyz_out__2006.csv',
           'df_abc_xyz_out__2005.csv', 'df_abc_xyz_out__2004.csv','df_abc_xyz_out__2003.csv']

list_year = [2014, 2013, 2012, 2011, 2010, 2009, 2008, 2007, 2006, 2005, 2004, 2003]
```

```python
df = pd.DataFrame()
```

```python
for i in range(0, len(list_csv)):
    df_abcxyz = pd.read_csv(list_csv[i], index_col=0)
    #df_abcxyz
    
    # взять год склеить их по Product ID по году
    dfyear = df4eda[df4eda['Year']==list_year[i]].merge(df_abcxyz, how = 'left', left_on = 'ProductID', right_on = 'ProductID')
#    print(dfyear['Year'].unique)
#    df = pd.concat([df, dfyear], axis=1, ignore_index=True)
    df = pd.concat([df, dfyear], ignore_index=True)
#    print(df.info())

```
