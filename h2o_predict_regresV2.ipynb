# Прогнозирование №2 - ПРОГНОЗ ПРИБЫЛИ ДЛЯ НОВЫХ ПРОДУКТОВ

```python
import pandas as pd
```

```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
df_new_product_templates = pd.DataFrame(
    {
        'Manufacturer' : ['Abbas', 'Barba', 'Pirum', 'Natura', 'VanArsdel', 'Salvus', 'Victoria'],
        'MonthNo' : [1, 1, 1, 1, 1, 1, 1],
        'Model_id_number' : ['33', '33', '33', '33', '33', '33', '33'],
        'CategoryInSegmentInCompete' : ['RuralProductivityNotCompete', 'RuralProductivityNotCompete', 'RuralProductivityNotCompete', 'RuralProductivityNotCompete', 'RuralProductivityNotCompete', 'RuralProductivityNotCompete', 'RuralProductivityNotCompete']
    }
)
df_new_product = df_new_product_templates.copy()
for k in range(2,13):
    df_df = df_new_product_templates.copy()
    df_df['MonthNo'] = k
    df_new_product = pd.concat([df_new_product, df_df], ignore_index=True)
#    print(df_new_product)
    
df_new_product
```

```python
df_new_product.to_csv('df_new_product.csv')
```

```python
df = h2o.H2OFrame(df_new_product)
df
```
