```python
import pandas as pd

file = 'df4eda.csv'
df = pd.read_csv(file, index_col=0)

# удалю Product ID, Product для кластеризации не нужен

# c годом так же распрощаюсь, интересует величина, которая будет предсказумеа и повторятся из года в год туда же
# и все связанное с датой - MonthName

#MfgisVanArsdel и isVanArsdel - уберу, слишком узкие и специфчные кажется

#City и State и Zip тоже уберу

# ManufacturerID_x, ManufacturerID_y - долой

# Category, IsCompeteHide, MonthID, Running Year - долой
df = df.drop(['ProductID', 'Year','Date', 'MonthIndex', 'Rolling Period Sort', 'Rolling Period',
             'Running Months', 'Month', 'Quarter', 'MfgisVanArsdel', 'isVanArsdel',
             'City', 'State', 'Product', 'ManufacturerID_x', 'Zip', 'Category', 'IsCompeteHide',
             'MonthID', 'RunningMonths', 'Running Year', 'ManufacturerID_y', 'MonthName'], axis=1)

# Units надо объединить, пусть будет один и больше одного
units_cat = {1:"one", 2:"moreone", 3:"moreone", 4:"moreone", 5:"moreone", 6:"moreone",
            7:"moreone", 8:"moreone", 9:"moreone", 10:"moreone", 11:"moreone", 12:"moreone"}
df['Units'] = df['Units'].map(units_cat)

#df['Units'].unique()
df
```
