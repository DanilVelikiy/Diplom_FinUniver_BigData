```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
#file = '/content/drive/MyDrive/Colab Notebooks/Diplom/df4eda2clean2.csv'
file = 'df4eda2clean2.csv'
df = h2o.import_file(file)
df = df.drop('C1')
df = df.drop('ProductID')
```

```python
df
```

```python
from h2o.estimators import H2OKMeansEstimator

# run h2o Kmeans to get good starting points
h2o_km = H2OKMeansEstimator(k=13, init="furthest", standardize=True)
h2o_km.train(training_frame=df)

user_points = h2o.H2OFrame(h2o_km.centers())

# show details
h2o_km.show()
```

```python
h2o_km.explain(df)
```
